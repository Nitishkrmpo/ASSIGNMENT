```python
df =sns.load_dataset('tips')
```

```python
df
```

```python
#numerical data plotting 
sns.relplot(x='total_bill', y ='tip', data=df )
```

```python
#by default the relplot give the scatter plot belation between the  y and x  axis
sns.scatterplot(x=df.total_bill, y=df.tip, data = df )
```

```python
#hue has differentiated between the smoker and non smoker in the dataset 
sns.scatterplot(x=df.total_bill, y=df.tip, data = df,hue = 'smoker' )
```

```python
#as the bills increase , the tips increases 
#highest tips was apid by the smoker 
sns.scatterplot(x='total_bill', y='tip', data = df,hue = 'smoker' )
```

```python
#relplot to make a line plot >> we generated a data
time = np.arange(500)
value=np.random.randn(500).cumsum()
df = pd.DataFrame({'time': time , 'value': value})
```
